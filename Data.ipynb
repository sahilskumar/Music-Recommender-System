```
import pandas as pd
df = pd.read_csv("output.csv")
df1 = pd.read_csv("output2.csv")
df.head()
df1.head()
```

```
new_df = df.drop_duplicates()
new_df1 = df1.drop_duplicates()
```

```
new_df.columns = ['lyrics1','lyrics2','similarity_score']
new_df1.columns = ['lyrics1','lyrics2','similarity_score']
```

```
frames = [new_df,new_df1]
result = pd.concat(frames)
result.head()
result.shape
```

```
result = result.drop_duplicates()
result.shape
```

```
result_negative_sample = result[result.similarity_score < 0.5]
result_negative_sample.shape
```

```
result_positive_sample = result[result.similarity_score >= 0.5]
result_positive_sample.shape
```

```
result_positive_sample.to_csv("result_positive_sample.csv",index=False)
result_negative_sample.to_csv("result_negative_sample.csv",index=False)
result.to_csv("result.csv",index=False)
```
